```python
import pandas as pd
import statistics

data = pd.read_csv('wine_cleared.csv')
data
```

```python
data['country'].value_counts()
```

```python
regex = r'\d{4}' # регулярное выражение для нахождения чисел
data['year'] = data['title'].str.findall(regex).str.get(0)
```

```python
data['is_usa'] = data['country'].apply(lambda x: 1 if x=='US' else 0)
data['is_france'] = data['country'].apply(lambda x: 1 if x=='France' else 0)
data['is_italy'] = data['country'].apply(lambda x: 1 if x=='Italy' else 0)
```

```python
data['is_italy'].sum()
```

```python
data['old_wine'] = data['year'].apply(lambda x: 1 if float(x)<2010 else 0)
data['old_wine'].sum()
```

```python
regex = r'\((.*?)\)'
data['locality'] = data['title'].str.findall(regex).str.get(0)
data['locality'].value_counts().head(10)
```

```python
country_population = pd.read_csv('country_population.zip', sep=';')
country_population[country_population['country']=='Italy']
```

```python
data.join(country_population.set_index('country'), on='country')
```

```python
country_area = pd.read_csv('country_area.zip', sep=';')
country_area
```

```python
data = data.join(country_area.rename(columns={'area':'area_country'}).set_index('country'), on='country')
```
